```python
#Maximillian_Darrington_Assingment1_1B
import pandas as pd
df = pd.read_excel('AMZN.xlsx', sheet_name='AMZN (1)',
                   index_col=0, parse_dates=True)
df.info()
# calculate % change and append as a new column
df["Daily Returns"] = df["Adj Close"].pct_change()

# Colors elements in a dataframe green(+), red(-)
def color_negative_red(value):
  if value < 0:
    color = 'red'
  elif value > 0:
    color = 'green'
  else:
    color = 'black'
  return 'color: %s' % color
# add styles: PctChange (pct w/2 decimal, color) and Volume (comma)
(df.style
    .applymap(color_negative_red, subset=['Daily Returns'])
    .format({'Pct Change': "{:.2%}"})
    .format({'Volume': "{:,}"}))
```

```python
df1=df['Daily Returns']

import matplotlib.pyplot as plt

x = df1
plt.hist(x,bins=10,color='r')
```
